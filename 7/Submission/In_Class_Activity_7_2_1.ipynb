## Part 4
* Create a pie chart showing the total number of times each Genre appears in the dataset.
* Find out the top 10 most prolific authors and plot their number of books as a bar plot
* BONUS: create a histogram showing the distribution of User Rating scores

```python
import matplotlib.pyplot as plt
import seaborn as sns

genre_count = dfbs['Genre'].value_counts()
plt.figure(figsize = (10,6))
plt.pie(genre_count, labels=genre_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Different Genres in total')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()

#bar chart

# Top 10 most prolific authors
top_authors = dfbs['Author'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_authors.index,top_authors.values,color = 'skyblue')
plt.title('10 Most Prolific Authors')
plt.xlabel('Authors')
plt.ylabel('No.of Books')
plt.xticks(rotation=45)
plt.show()

#histogram

user_ratings = dfbs['User Rating']
print(user_ratings)
plt.figure(figsize=(10,6))
plt.hist(user_ratings, bins=10, color='skyblue', edgecolor='black')
plt.title('user rating distribution')
plt.xlabel('User Rating')
plt.ylabel('values')
plt.show()
```
