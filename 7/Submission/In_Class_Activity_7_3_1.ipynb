```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dfdb = pd.read_csv('/content/diabetes.csv')

corrl_matrix = dfdb.corr()

plt.figure(figsize=(10, 6))

sns.heatmap(corrl_matrix, annot=True, fmt=".2f", cmap="viridis",
            square=True, linewidths=0.5, linecolor='white',
            xticklabels=corrl_matrix.columns, yticklabels=corrl_matrix.columns)

plt.title('Correlation Heatmap')
plt.show()
```
